### 5.1) Whether the average response time across complaint types is similar or not (overall)




```python
from scipy.stats import f_oneway
```

```python
new_data['Complaint Type'].unique()
```

```python
data=new_data['Complaint Type']
value=new_data['Request_Closing_Time']
```

```python
location1= data== 'Noise - Street/Sidewalk'
location2= data== 'Blocked Driveway'
location3= data== 'Illegal Parking'
location1
```

```python
new_location1=value[location1].dropna()
new_location2=value[location2].dropna()
new_location3=value[location3].dropna()
new_location1
```

### Null hypothesis: The average response time across complaint types is similar.
### Alternative hypothesis: The average response time across complaint types is NOT similar.

```python
print(f_oneway(new_location1,new_location2,new_location3))
```

```python
print('pvalue is greater than 0.05 so it significe we the REJECT NULL HYPOTHESIS.')
print('The average response time across complaint types is NOT similar.')
```

### 5.2) Are the type of complaint or service requested and location related?

```python
frequency_table=pd.crosstab(new_data['Complaint Type'],new_data['Location Type'])
frequency_table
```

```python
alpha=0.05 
```

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected= chi2_contingency(frequency_table)
print('p value is :-' "{:.5f}".format(p))
```

### Null hypothesis:- There is NO relation between 'Complaint Type' & 'Location Type' categorical variables.
### Alternative hypothesis:- There is a relation between 'Complaint Type' & 'Location Type' categorical variables.

```python
if p<= alpha:
    print("Accept the null hypothesis, There is NO relation between 'Complaint Type' & 'Location Type' categorical variables.")
else:
    print("Reject the null hypothesis, There is a relation between 'Complaint Type' & 'Location Type' categorical variables." )
```
